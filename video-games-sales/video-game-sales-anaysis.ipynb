# <p style="background-color:IndianRed;font-family:Verdana;font-size:135%;color:lightgray;text-align:center;border-radius:50px 5px;">Loading The Dataset</p>


Let’s load the dataset and index the game names in the dataset. For this, let’s use the index_col argument and assign the dataset to the games variable.

```python
games = pd.read_csv("https://raw.githubusercontent.com/diogo-costa-silva/datasets/main/video-game-sales.csv", index_col = "Name")
```


This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

Fields include: 

- Rank - Ranking of overall sales

- Name - The games name

- Platform - Platform of the games release (i.e. PC,PS4, etc.)

- Year - Year of the game's release

- Genre - Genre of the game

- Publisher - Publisher of the game

- NA_Sales - Sales in North America (in millions)

- EU_Sales - Sales in Europe (in millions)

- JP_Sales - Sales in Japan (in millions)

- Other_Sales - Sales in the rest of the world (in millions)

- Global_Sales - Total worldwide sales (in millions)

# <p style="background-color:IndianRed;font-family:Verdana;font-size:135%;color:lightgray;text-align:center;border-radius:50px 5px;">Data Visualization</p>

```python
# set a graphic style and a figure size with Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'darkgrid')
sns.set(rc={'figure.figsize':(12,9)})
```

Let's draw a box plot for the `Year` column.

```python
# Filter out rows with 'Year' value of 0
filtered_games = games[games["Year"] != 0]

# Create the boxplot
filtered_games["Year"].plot.box()
plt.title('Boxplot of Game Release Years')
plt.ylabel('Year')
plt.show()
```

Let's draw a histogram for the `Year` column.

```python
filtered_games["Year"].hist(bins=20)
```

As you can see, most of the games were released around 2010.

```python
# 1. Distribution of Game Release Years
plt.figure(figsize=(12, 6))
sns.histplot(data=filtered_games, x='Year', bins=30, color='royalblue', kde=True)
plt.title('Distribution of Game Release Years')
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.show()
```

```python
# Reset the index temporarily for the visualization
temp_games = filtered_games.reset_index()

# Distribution of Game Release Years
plt.figure(figsize=(12, 6))
sns.histplot(data=temp_games, x='Year', bins=30, color='royalblue', kde=True)
plt.title('Distribution of Game Release Years')
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.show()
```

```python
# 2. Top 10 Game Publishers by Number of Games
top_publishers = filtered_games['Publisher'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_publishers.values, y=top_publishers.index, palette="viridis")
plt.title('Top 10 Game Publishers by Number of Games')
plt.xlabel('Number of Games')
plt.ylabel('Publisher')
plt.show()
```

```python
# 3. Global Sales by Genre
genre_sales = filtered_games.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=genre_sales.values, y=genre_sales.index, palette="magma")
plt.title('Total Global Sales by Genre')
plt.xlabel('Total Sales (in millions)')
plt.ylabel('Genre')
plt.show()
```

```python
# 4. Sales by Platform
platform_sales = filtered_games.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=platform_sales.values, y=platform_sales.index, palette="coolwarm")
plt.title('Total Global Sales by Platform')
plt.xlabel('Total Sales (in millions)')
plt.ylabel('Platform')
plt.show()
```

```python
# 5. Yearly Global Sales Trend
yearly_sales = filtered_games.groupby('Year')['Global_Sales'].sum()

plt.figure(figsize=(14, 7))
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, color="purple", marker="o")
plt.title('Yearly Global Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales (in millions)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# 6. Top Genres by Region

# Sales by genre in North America
na_sales = filtered_games.groupby('Genre')['NA_Sales'].sum().sort_values(ascending=False).head(5)

# Sales by genre in Europe
eu_sales = filtered_games.groupby('Genre')['EU_Sales'].sum().sort_values(ascending=False).head(5)

# Sales by genre in Japan
jp_sales = filtered_games.groupby('Genre')['JP_Sales'].sum().sort_values(ascending=False).head(5)

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))

sns.barplot(x=na_sales.values, y=na_sales.index, palette="Blues_r", ax=axes[0])
axes[0].set_title('Top Genres by Sales in North America')
axes[0].set_xlabel('Sales (in millions)')
axes[0].set_ylabel('Genre')

sns.barplot(x=eu_sales.values, y=eu_sales.index, palette="Greens_r", ax=axes[1])
axes[1].set_title('Top Genres by Sales in Europe')
axes[1].set_xlabel('Sales (in millions)')
axes[1].set_ylabel('Genre')

sns.barplot(x=jp_sales.values, y=jp_sales.index, palette="Reds_r", ax=axes[2])
axes[2].set_title('Top Genres by Sales in Japan')
axes[2].set_xlabel('Sales (in millions)')
axes[2].set_ylabel('Genre')

plt.tight_layout()
plt.show()
```

```python
# 7. Correlation Heatmap
correlation_matrix = filtered_games[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Sales in Different Regions')
plt.show()
```

Here are some insights from the heatmap:

North American (NA) sales and European (EU) sales have a high correlation, indicating that games that sell well in NA often also sell well in Europe.
Both NA and EU sales have a strong correlation with Global sales, which is expected since these two regions contribute significantly to the global market.
Japanese (JP) sales have a weaker correlation with NA and EU sales, suggesting that the Japanese market might have distinct preferences compared to the western markets.


```python
# Combine the sales data for the top genres by region
combined_data = pd.DataFrame({
    'North America': na_sales,
    'Europe': eu_sales,
    'Japan': jp_sales
}).fillna(0)  # Fill missing values with 0

# Plotting the combined bar chart
plt.figure(figsize=(14, 8))
combined_data.plot(kind='bar', figsize=(14, 8), colormap="viridis", width=0.8)
plt.title('Top Genres by Sales in North America, Europe, and Japan')
plt.xlabel('Genre')
plt.ylabel('Sales (in millions)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# Transpose the combined data for plotting
transposed_data = combined_data.transpose()

# Plotting the combined bar chart with regions on the x-axis
plt.figure(figsize=(14, 8))
transposed_data.plot(kind='bar', figsize=(14, 8), colormap="viridis", width=0.8)
plt.title('Top Genres by Sales in Different Regions')
plt.xlabel('Region')
plt.ylabel('Sales (in millions)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 9. Sales Distribution by Platform
platform_global_sales = filtered_games.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=platform_global_sales.values, y=platform_global_sales.index, palette="cubehelix")
plt.title('Top 10 Platforms by Global Sales')
plt.xlabel('Total Sales (in millions)')
plt.ylabel('Platform')
plt.show()
```

```python
# 10. Genre Popularity Over Time

# Selecting top 5 genres by global sales
top_genres = filtered_games.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(5).index

# Filter data to only include these top genres
top_genre_data = filtered_games[filtered_games['Genre'].isin(top_genres)]

# Group data by year and genre to get yearly sales
yearly_genre_sales = top_genre_data.groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack()

plt.figure(figsize=(15, 8))
sns.lineplot(data=yearly_genre_sales, dashes=False, palette="husl", markers=True)
plt.title('Genre Popularity Over Time')
plt.xlabel('Year')
plt.ylabel('Total Sales (in millions)')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.legend(title="Genre")
plt.tight_layout()
plt.show()
```

# <p style="background-color:IndianRed;font-family:Verdana;font-size:135%;color:lightgray;text-align:center;border-radius:50px 5px;">Filtering</p>

The `filter` method allows you to filter the DataFrame for rows and columns. The returned DataFrame contains only the rows and columns specified by the method. Let’s see the rows whose genre is action.

```python
games[games["Genre"] == "Action"]
```

You can also filter with a variable. Let's show this.

```python
games_by_genre = games["Genre"] == "Action"
games[games_by_genre].head()
```

Let’s create a variable for games in 2010 and then filter by the two criteria.

```python
games_in_2010 = games["Year"] == 2010
games[games_by_genre & games_in_2010]
```

You can also see games in 2010 or the action genre with a pipe.

```python
games[games_by_genre | games_in_2010]
```

You can also see values greater or less than a certain value. First, let’s create a variable named after_2015 and filter by this variable.

```python
after_2015 = games["Year"] > 2015
games[after_2015]
```

You can see the games between 2000 and 2010 with the `between` method.

```python
mid_2000s = games["Year"].between(2000, 2010)
games[mid_2000s]
```

You can also filter by indexes. Let’s see the game names with the `sport` in them. Keep in mind that the `str` method is used to use string methods.

```python
sport_in_title = games.index.str.lower().str.contains("sport")
games[sport_in_title]
```

You can find the statistics of the numeric columns in the dataset. Let’s find the mean of global sales.

```python
games["Global_Sales"].mean()
```

# <p style="background-color:IndianRed;font-family:Verdana;font-size:135%;color:lightgray;text-align:center;border-radius:50px 5px;">Grouping</p>


The `groupby` method allows you to group the data according to the categories and apply a function to the categories. To show this, let’s first create the group object for the `Genre` column and apply the `sum` method to the `Global_Sales.

```python
genres = games.groupby("Genre")
genres["Global_Sales"].sum()
```

You can also sort the values in descending order with the `sort_values` method.

```python
genres["Global_Sales"].sum().sort_values(ascending = False)
```

Let's take a look at the mean of global sales according to the genre column with the `groupby` method.

```python
games_avg=games["Global_Sales"].groupby(games["Genre"]).mean()
games_avg
```

Let's draw a pie plot for the global sales.

```python
games_avg.plot.pie()
```

As you can see, the platform genre has the most sales.

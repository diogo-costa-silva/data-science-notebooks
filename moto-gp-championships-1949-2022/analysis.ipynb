## Individual Dataset Analysis

Step 1: Basic Statistics, Null Values, and Unique Counts
We'll begin by examining:

Basic statistics for numerical columns.
Count of null values in each column.
Count of unique values in each column.

```python
# Function to perform basic analysis on a dataset
def basic_analysis(df, name):
    print(f"--- {name} Dataset Analysis ---")
    print("Basic Statistics:")
    display(df.describe())
    print("Null Values:")
    display(df.isnull().sum())
    print("Unique Counts:")
    display(df.nunique())
    print('-'*50)

```

```python
# Perform basic analysis on the Constructors dataset
basic_analysis(df_constructors, "Constructors")
```

The dataset has data from 1949 to 2021.

- Null Values:

- Unique Counts:
